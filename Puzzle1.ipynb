```python
import pandas as pd
```

```python
df1 = pd.read_csv('DepartmentData.csv')
df2 = pd.read_csv('EmployeeData.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3 = pd.merge(df1, df2, on='department_id')
df3
```

```python
df3["total_salary"] = df3["base_salary"] + df3["bonus"]
df3
```

```python
df3['total_salary'] = df3['total_salary']-(df3['total_salary']*df3['tax_rate'])
average_spectral_salary = df3.groupby("department_name")['total_salary'].mean()
print(average_spectral_salary.sort_values(ascending = False))
```
